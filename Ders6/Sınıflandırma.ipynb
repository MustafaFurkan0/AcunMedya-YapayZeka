```python
df = pd.read_csv("spam_ham_dataset.csv")
df.head()
```

```python
df = df[ ['text','label_num'] ]
df.head(4)
```

```python
import nltk
nltk.download("stopwords")
```

```python
# hediye? HEDiye!# HeDiye Hediye H3DİYE => bu şekilde istemiyoruz hepsini numeric yapıcaz

import string

ps = nltk.stem.porter.PorterStemmer()

def clean_text(text):
    text = text.lower() # Bütün harfleri küçük yap
    
    text = text.translate(str.maketrans('','',string.punctuation)) # Bütün noktalama işaretleri kaldır
    
    # "Merhaba,Dünya." =>> ["merhaba","dünya"]
    words = text.split()
    
    words = [ word for word in words if not word.isdigit()] # => Yalnızca numeric kelimeler alınmayacak
    # Stem -> Kelimeleri köküne indirgemek -> Running,Run -> Run
    # Remove Stpwords -> The, is, at, on
    words = [ ps.stem(word) for word in words if word not in nltk.corpus.stopwords.words('english')]

    return " ".join(words)
```

```python
df.shape
```

```python
df['clean_text'] = df['text'].apply(clean_text)
```

```python
df.head(4)
```

```python
from sklearn.model_selection import train_test_split

X = df['clean_text']
y = df['label_num']
```

```python
# Vectorize -> metni numerize etmek
from sklearn.feature_extraction.text import TfidfVectorizer # TFIDF vectorizer nasıl çalışır?
```

```python
tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df['clean_text'])
X_tfidf
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size = 0.3, random_state=42)
```

```python
# Naive Bayes -> Bayes Teoremi
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
# precision -> kesinlik
# recall -> duyarlılık

# f1 score -> ikisinin karşılaştırması 1 e yaklaştıkça doğrudur

# support -> veri sayısı
```
