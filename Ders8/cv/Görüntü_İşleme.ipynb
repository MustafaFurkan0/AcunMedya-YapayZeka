# OCR (Optical Character Recognition – Optik Karakter Tanıma) 
Görsellerde veya taranmış belgelerde yer alan yazılı metinleri dijital ve düzenlenebilir metne dönüştürme işlemidir. 
Özellikle taranmış evraklar, fatura, kitap sayfaları veya el yazısı notlar gibi görsel verilerdeki yazıyı makine tarafından okunabilir hale getirmek için kullanılır.

```python
import pytesseract
import os
from PIL import Image, ImageEnhance, ImageFilter
```

```python
# Manuel olarak tessdata yolunu belirt
os.environ['TESSDATA_PREFIX'] = r'C:\Program Files\Tesseract-OCR\tessdata'

# Tesseract yolunu belirt
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
image = Image.open('dataset/images/6.jpg')
```

```python
metin = pytesseract.image_to_string(image, lang="tur", config="--psm 6")
if not metin.strip():
    print("OCR çıktı vermedi. Görüntü kalitesi, dil dosyası veya konfigürasyon sorun olabilir.")
else:
    print(metin)
```
