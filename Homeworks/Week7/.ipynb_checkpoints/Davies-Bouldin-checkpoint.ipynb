# Davies-Bouldin Index (DBI)
Davies-Bouldin Index (DBI), kümeleme algoritmalarının performansını değerlendirmek için kullanılan bir metriğe dayanır. Daha düşük DBI değeri, kümelerin birbirinden daha iyi ayrıldığını ve daha sıkı bir şekilde gruplanmış olduğunu gösterir. Bu metot, özellikle KMeans, DBSCAN gibi kümeleme algoritmalarını değerlendirirken kullanılır.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
X = df[ ["Annual Income (k$)", "Spending Score (1-100)"] ].values # df yi arraya çevir
```

```python
model = KMeans(n_clusters=5, random_state=42)
model.fit(X)
```

```python
y = model.predict(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], c="red", s=500, marker="X")
plt.title("KMeans Kümeleme ve Merkezler")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
# Davies-Bouldin skorunu hesapla
db_score = davies_bouldin_score(X, y)
print(f"Davies-Bouldin Index Skoru: {db_score:.3f}")
# Skor ne kadar düşükse, kümeler o kadar iyi ayrışmış ve sıkı gruplanmıştır.
```
