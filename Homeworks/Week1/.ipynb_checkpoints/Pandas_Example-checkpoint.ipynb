```python
import pandas as pd
```

## 1) read_csv() => CSV dosyasını okur.

```python
df1 = pd.read_csv("prog_book.csv")
df1
```

## 2) duplicated() =>  Yinelenen satırları kontrol eder.

```python
df1.duplicated()
```

## 3) info)() => Veri çerçevesi hakkında bilgi verir.

```python
df1.info()
```

## 4) describe() => Temel istatistiksel özet sağlar.

```python
df1.describe()
```

## 5) sample() => Rastgele satır örnekleri seçer.

```python
df1.sample()
```

## 6) rename() => Sütun veya indeks adlarını değiştirir.

```python
df1.rename(columns={'Number_Of_Pages' : 'Sayfa_Sayısı'})
```

## 7) pivot_table(): Özet tablo oluşturur.

```python
pivot = pd.pivot_table(df1, values='Reviews', index='Type', columns='Book_title', aggfunc='sum', margins=True)
#print(pivot)
```

## 8) rank() => Verileri sıralar ve sıralama pozisyonu döner.

```python
df1.rank()
```

## 9) sort_values() => Verileri belirli bir sütuna göre sıralar

```python
df1.sort_values(by = 'Rating', ascending=False )
```

## 10) value_counts() => Bir sütundaki her bir benzersiz değerin kaç kere tekrar ettiğini sayar.

```python
df1['Type'].value_counts()
```
