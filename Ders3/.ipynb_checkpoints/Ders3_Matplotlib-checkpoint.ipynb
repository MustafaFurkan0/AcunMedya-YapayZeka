```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [100,200,300,400,500]
x,y
```

```python
plt.plot(x,y , color = "green", linestyle = "dashed", marker = "s", label = "Satış")
plt.legend() 
plt.grid()
plt.xlabel("Günler")
plt.ylabel("Satışlar")
plt.show()
```

```python
import numpy as np
```

```python
x= np.linspace(0,10,100) # 0 ile 10 arasında 100 adet eşit aralıklı sayı üret
x
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x, y1, label = "sin(x)", color = "red")
plt.plot(x, y2, label = "cos(x)", color = "blue")
plt.legend()

plt.ylabel("Sin, Cos Değerleri")
plt.xlabel("X Değeri")

plt.show()
```
